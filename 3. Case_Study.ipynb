## Import modules

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'User_ID': range(1, 101),
    'Age_Group': ['18-24']*30 + ['25-34']*40 + ['35-44']*30,
    'Visitor_Type': ['New']*50 + ['Returning']*50, 
    'Avg_Time_Spent': [3, 5, 7, 2, 4, 6]*16 + [5, 6, 4, 7]
}
```

```python
user_data = pd.DataFrame(data)
user_data.head()
```

```python
age_group_segment = user_data.groupby('Age_Group')['User_ID'].count()
visitor_type_segment = user_data.groupby('Visitor_Type')['User_ID'].count()

print("Users by Age Group:")
print(age_group_segment)
print("\nUsers by Visitor Type:")
print(visitor_type_segment)
```

### Bar chart for Age Group segmentation

```python
age_group_segment.plot(kind='bar', color='lightblue', title='User Segmentation by Age Group')
plt.ylabel('Number of Users')
plt.xlabel('Age Group')
plt.show()
```

### Pie chart for Visitor Type segmentation

```python
visitor_type_segment.plot(kind='pie', autopct='%1.1f%%', title='User Segmentation by Visitor Type', startangle=140)
plt.ylabel('')
plt.show()
```
